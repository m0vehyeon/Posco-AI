# Histogram of oriented gradient (HOG)

```python
import matplotlib.pyplot as plt
import skimage.feature
import skimage.data
import cv2

#img = cv2.imread('picture/logo.png')
#    
# https://scikit-image.org/docs/dev/api/skimage.data.html
image = skimage.data.astronaut()
plt.imshow(image)
plt.title('Input image')
plt.show()
```

### HOG feature image 

**hog(image, orientations, pixels_per_cell, cells_per_block, visualize)**  
Parameters:	
- image  Input image. Array.
- orientations  Numer of orientation bins.
- pixels_per_cell  Size (in pixels) of a cell.
- cells_per_block  Number of cells in each block.
- visualize  Also return an image of the HOG. Bool.

```python
hog_feature, hog_image = skimage.feature.hog(image, orientations=9, pixels_per_cell=(8, 8),cells_per_block=(2, 2),
                                             visualize=True)
```

```python
print('image : ',image.shape)
print('hog_feature : ',hog_feature.shape)
print('hog_image : ',hog_image.shape)
```

```python
plt.figure(figsize=(12,10))
plt.imshow(hog_image,cmap=plt.cm.gray)
plt.title('HOG')
plt.show()
```
